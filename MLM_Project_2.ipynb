# **Analysis of Non-Normally Distributed Variables and Scaling**

Identifying Non-Normality

The non-categorical variables 'Quantity', 'Value', and 'Weight' were assessed for normality using the Shapiro-Wilk test. This test helps determine if a dataset follows a normal distribution.

Results of Normality Test

The Shapiro-Wilk test indicated that all three variables — Quantity, Value, and Weight — deviated significantly from a normal distribution (p-value < 0.05 for each). This suggests the presence of skewness or other departures from normality in these variables.

Addressing Non-Normality: Scaling with MinMaxScaler

To handle the non-normality and optimize model performance, a scaling transformation was applied. Specifically, the MinMaxScaler was used to scale the non-normally distributed variables.

How MinMaxScaler Works

MinMaxScaler transforms the data by mapping it to a specific range, typically between 0 and 1. This scaling method preserves the original distribution of the data while ensuring all features are on a comparable scale. The formula for MinMax scaling is as follows:

X_scaled = (X - X_min) / (X_max - X_min)

Where:

* X_scaled is the scaled value
* X is the original value
* X_min is the minimum value of the feature
* X_max is the maximum value of the feature

Benefits of Scaling

Scaling the non-normally distributed variables using MinMaxScaler provides several advantages:

* Equal Treatment of Variables: It ensures that features with larger values do not disproportionately influence the model's learning process.
* Improved Model Performance: Scaling can enhance the performance of machine learning models, particularly those sensitive to feature scaling, such as linear regression and k-means clustering.
* Enhanced Model Stability: It can contribute to the stability and robustness of the models, making them less susceptible to outliers and extreme values.

Impact on Machine Learning Models

By addressing the non-normality of the variables through scaling, we aim to create a more favorable input for the machine learning models. This transformation can potentially lead to improved model accuracy, interpretability, and overall performance.

* **Variables:**
    * `Shipping_Method`: Categorical with three levels (0.0 - Air, 1.0 - Land, 2.0 - Sea)

**Hypotheses:**

* **Null Hypothesis (H0):** There is no significant association between `Shipping_Method` and `Import_Export`.
* **Alternative Hypothesis (H1):** There is a significant association between `Shipping_Method` and `Import_Export`.

**Statistical Test:**

* Chi-squared test of independence was used to assess the association between the two categorical variables.

**Results:**

| Statistic | Value |
|---|---|
| Chi-squared Statistic | 3.737 |
| Degrees of Freedom | 2 |
| P-value | 0.154 |

**Interpretation:**

The p-value (0.154) is greater than the commonly used significance level of 0.05. Therefore, we **fail to reject the null hypothesis**.

**Conclusion:**

This analysis suggests that there is **no statistically significant association** between `Shipping_Method` and `Import_Export` based on the observed data. This implies that the choice of shipping method is likely independent of whether a transaction is an import or export.

# **Data Preprocessing**

```python
# Handling Missing Data
def missing_data_summary(df):
    """
    Creates a DataFrame summarizing missing data statistics.

    Args:
        df: The input DataFrame.

    Returns:
        A DataFrame with missing data count and percentage for each variable and row.
    """
    # Column-wise summary
    missing_count_col = df.isnull().sum()
    missing_percentage_col = (missing_count_col / len(df)) * 100
    missing_summary_col = pd.DataFrame({
        'Missing Count': missing_count_col,
        'Missing Percentage': missing_percentage_col
    })
    # Row-wise summary
    missing_count_row = df.isnull().sum(axis=1)
    missing_percentage_row = (missing_count_row / len(df.columns)) * 100
    missing_summary_row = pd.DataFrame({
        'Missing Count': missing_count_row,
        'Missing Percentage': missing_percentage_row
    }, index=df.index)  # Use original DataFrame index

    return missing_summary_col, missing_summary_row

# Calculate missing data summary for categorical subset
sg39_missing_summary_cat_col, sg39_missing_summary_cat_row = missing_data_summary(sg39_ds_cat)

# Calculate missing data summary for non-categorical subset
sg39_missing_summary_noncat_col, sg39_missing_summary_noncat_row = missing_data_summary(sg39_ds_noncat)

# Drop columns and rows with more than 50% missing data for categorical subset
sg39_ds_cat_treated = sg39_ds_cat.dropna(thresh=len(sg39_ds_cat) * 0.5, axis=1)
sg39_ds_cat_treated = sg39_ds_cat_treated.dropna(thresh=len(sg39_ds_cat_treated.columns) * 0.5, axis=0)

# Drop columns and rows with more than 50% missing data for non-categorical subset
sg39_ds_noncat_treated = sg39_ds_noncat.dropna(thresh=len(sg39_ds_noncat) * 0.5, axis=1)
sg39_ds_noncat_treated = sg39_ds_noncat_treated.dropna(thresh=len(sg39_ds_noncat_treated.columns) * 0.5, axis=0)

# Calculate total variables and rows dropped
sg39_dropped_variables_cat = len(sg39_ds_cat.columns) - len(sg39_ds_cat_treated.columns)
sg39_dropped_rows_cat = len(sg39_ds_cat) - len(sg39_ds_cat_treated)
sg39_dropped_variables_noncat = len(sg39_ds_noncat.columns) - len(sg39_ds_noncat_treated.columns)
sg39_dropped_rows_noncat = len(sg39_ds_noncat) - len(sg39_ds_noncat_treated)

sg39_total_dropped_variables = sg39_dropped_variables_cat + sg39_dropped_variables_noncat
sg39_total_dropped_rows = sg39_dropped_rows_cat + sg39_dropped_rows_noncat

# Print missing data summaries
print("Categorical Missing Data Summary (Columns):")
print(sg39_missing_summary_cat_col)
print("\nCategorical Missing Data Summary (Rows):")
print(sg39_missing_summary_cat_row)
print("\nNon-Categorical Missing Data Summary (Columns):")
print(sg39_missing_summary_noncat_col)
print("\nNon-Categorical Missing Data Summary (Rows):")
print(sg39_missing_summary_noncat_row)

# Print total dropped variables and rows
print(f"\nTotal variables dropped: {sg39_total_dropped_variables}")
print(f"Total rows dropped: {sg39_total_dropped_rows}")
```

# **Descriptive Statistics**

```python
# Function to calculate descriptive statistics for categorical variables
def sg39_categorical_stats(df, cat_columns):
    sg39_results = []

    for sg39_col in cat_columns:
        sg39_counts = df[sg39_col].value_counts()
        sg39_proportions = sg39_counts / len(df)
        sg39_mode = sg39_counts.idxmax()
        sg39_min = df[sg39_col].min()
        sg39_max = df[sg39_col].max()

        for category, count in sg39_counts.items():
            sg39_results.append({
                "Column": sg39_col,
                "Category": category,
                "Frequency": count,
                "Proportion": sg39_proportions[category],
                "Minimum": sg39_min,
                "Maximum": sg39_max,
                "Mode": sg39_mode
            })

    return pd.DataFrame(sg39_results)

# Function to calculate correlation for categorical columns
def sg39_categorical_correlation(df, cat_columns, method='spearman'):
    sg39_corr_results = pd.DataFrame(index=cat_columns, columns=cat_columns, dtype=float)

    for i in range(len(cat_columns)):
        for j in range(i, len(cat_columns)):
            sg39_col1, sg39_col2 = cat_columns[i], cat_columns[j]
            if i == j:
                # Perfect correlation for the same variable
                corr = 1.0
            else:
                if method == 'spearman':
                    corr, _ = spearmanr(df[sg39_col1].astype('category').cat.codes,
                                        df[sg39_col2].astype('category').cat.codes)
                elif method == 'kendall':
                    corr, _ = kendalltau(df[sg39_col1].astype('category').cat.codes,
                                         df[sg39_col2].astype('category').cat.codes)
            sg39_corr_results.loc[sg39_col1, sg39_col2] = corr
            sg39_corr_results.loc[sg39_col2, sg39_col1] = corr

    return sg39_corr_results

# Retrieve categorical columns from sg39_ds_cat_encoded
sg39_cat_columns = sg39_ds_cat_encoded.columns.tolist()

# Generate descriptive stats and correlations for these columns
sg39_categorical_stats_df = sg39_categorical_stats(sg39_ds_preprocessed, sg39_cat_columns)
sg39_spearman_corr_df = sg39_categorical_correlation(sg39_ds_preprocessed, sg39_cat_columns, method='spearman')
sg39_kendall_corr_df = sg39_categorical_correlation(sg39_ds_preprocessed, sg39_cat_columns, method='kendall')

# Replace NaN correlations with 1 for diagonal elements (if any remain)
sg39_spearman_corr_df.fillna(1, inplace=True)
sg39_kendall_corr_df.fillna(1, inplace=True)

# Display results
print("Descriptive Statistics:")
print(sg39_categorical_stats_df)  # Show all rows

print("\nSpearman Correlation:")
print(sg39_spearman_corr_df)

print("\nKendall Correlation:")
print(sg39_kendall_corr_df)
```

```python
# Function to calculate descriptive statistics for numerical variables
def sg39_numerical_stats(df, num_columns):
    sg39_results = []

    for sg39_col in num_columns:
        sg39_data = df[sg39_col].dropna()  # Exclude NaN values
        sg39_min = sg39_data.min()
        sg39_max = sg39_data.max()
        sg39_mean = sg39_data.mean()
        sg39_median = sg39_data.median()
        sg39_mode = sg39_data.mode().iloc[0] if not sg39_data.mode().empty else np.nan
        sg39_range = sg39_max - sg39_min
        sg39_std_dev = sg39_data.std()
        sg39_skewness = skew(sg39_data)
        sg39_kurt = kurtosis(sg39_data)

        # Coefficient of Variation (CV)
        sg39_cv = sg39_std_dev / sg39_mean if sg39_mean != 0 else np.nan

        # Confidence Interval (95%)
        sg39_conf_interval = norm.interval(0.95, loc=sg39_mean, scale=sg39_std_dev / np.sqrt(len(sg39_data)))

        # Add results
        sg39_results.append({
            "Column": sg39_col,
            "Minimum": sg39_min,
            "Maximum": sg39_max,
            "Mean": sg39_mean,
            "Median": sg39_median,
            "Mode": sg39_mode,
            "Range": sg39_range,
            "Standard Deviation": sg39_std_dev,
            "Skewness": sg39_skewness,
            "Kurtosis": sg39_kurt,
            "Coefficient of Variation": sg39_cv,
            "Confidence Interval (95%)": sg39_conf_interval
        })

    return pd.DataFrame(sg39_results)

# Function to calculate correlation for numerical variables
def sg39_numerical_correlation(df, num_columns, method='pearson'):
    sg39_corr_results = pd.DataFrame(index=num_columns, columns=num_columns, dtype=float)

    for i in range(len(num_columns)):
        for j in range(i, len(num_columns)):
            sg39_col1, sg39_col2 = num_columns[i], num_columns[j]
            if i == j:
                corr = 1.0  # Perfect correlation for the same variable
            else:
                if method == 'pearson':
                    corr, _ = pearsonr(df[sg39_col1].dropna(), df[sg39_col2].dropna())
                elif method == 'spearman':
                    corr, _ = spearmanr(df[sg39_col1].dropna(), df[sg39_col2].dropna())
            sg39_corr_results.loc[sg39_col1, sg39_col2] = corr
            sg39_corr_results.loc[sg39_col2, sg39_col1] = corr

    return sg39_corr_results

# Retrieve numerical columns from sg39_ds_noncat_scaled
sg39_num_columns = sg39_ds_noncat_scaled.columns.tolist()

# Generate statistics and correlations for these columns
sg39_numerical_stats_df = sg39_numerical_stats(sg39_ds_noncat_scaled, sg39_num_columns)
sg39_pearson_corr_df = sg39_numerical_correlation(sg39_ds_noncat_scaled, sg39_num_columns, method='pearson')
sg39_spearman_corr_df = sg39_numerical_correlation(sg39_ds_noncat_scaled, sg39_num_columns, method='spearman')

# Replace NaN correlations with 1 for diagonal elements (if any remain)
sg39_pearson_corr_df.fillna(1, inplace=True)
sg39_spearman_corr_df.fillna(1, inplace=True)

# Display results
print("Numerical Statistics:")
print(sg39_numerical_stats_df)

print("\nPearson Correlation:")
print(sg39_pearson_corr_df)

print("\nSpearman Correlation:")
print(sg39_spearman_corr_df)
```

```python
# Assuming your preprocessed dataset is named 'sg39_ds_preprocessed'
# and your non-categorical columns are 'Quantity', 'Value', 'Weight'

# List of non-categorical variables
sg39_non_categorical_vars = ['Quantity', 'Value', 'Weight']

# Loop through each variable
for sg39_var in sg39_non_categorical_vars:
    print(f"\n--- Tests for {sg39_var} ---")

    # Shapiro-Wilk Test
    sg39_shapiro_stat, sg39_shapiro_p_value = stats.shapiro(sg39_ds_preprocessed[sg39_var])
    print(f"Shapiro-Wilk Test - Statistic: {sg39_shapiro_stat}, P-value: {sg39_shapiro_p_value}")
    if sg39_shapiro_p_value < 0.05:
        print(f"Data does not look normal (reject H0) at 5% significance level")
    else:
        print(f"Data looks normal (fail to reject H0) at 5% significance level")

    # Kolmogorov-Smirnov Test
    sg39_ks_stat, sg39_ks_p_value = stats.kstest(sg39_ds_preprocessed[sg39_var], 'norm')
    print(f"Kolmogorov-Smirnov Test - Statistic: {sg39_ks_stat}, P-value: {sg39_ks_p_value}")
    if sg39_ks_p_value < 0.05:
        print(f"Data does not look normal (reject H0) at 5% significance level")
    else:
        print(f"Data looks normal (fail to reject H0) at 5% significance level")

    # Anderson-Darling Test
    sg39_anderson_result = stats.anderson(sg39_ds_preprocessed[sg39_var])
    print("Anderson-Darling Test - Statistic:", sg39_anderson_result.statistic)
    # Check only for 5% significance level
    if sg39_anderson_result.statistic < sg39_anderson_result.critical_values[2]:  # Index 2 corresponds to 5%
        print(f"Data looks normal (fail to reject H0) at 5% significance level")
    else:
        print(f"Data does not look normal (reject H0) at 5% significance level")

    # Jarque-Bera Test
    sg39_jb_stat, sg39_jb_p_value = stats.jarque_bera(sg39_ds_preprocessed[sg39_var])
    print(f"Jarque-Bera Test - Statistic: {sg39_jb_stat}, P-value: {sg39_jb_p_value}")
    if sg39_jb_p_value < 0.05:
        print(f"Data does not look normal (reject H0) at 5% significance level")
    else:
        print(f"Data looks normal (fail to reject H0) at 5% significance level")
```
